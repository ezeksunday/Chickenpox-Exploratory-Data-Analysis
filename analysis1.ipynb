# Visualizing trend using Line plot

```python
sns.set_theme(style="whitegrid")
fig,ax=plt.subplots(figsize=(19,7))
sns.lineplot(data=df,x=df.index.values,y=df.sum(axis=1),ax=ax)
sns.scatterplot(data=df,x=df.index.values,y=df.sum(axis=1),marker='^',s=300,color='orange',ax=ax)
plt.ylabel('Number of Chickenpox cases  per year')
plt.xlabel('YEARS')
plt.title('Visualizing the Trend Of chicken pox cases over the years (2005 to 2014)')
```

# Visualizing trend using Bar plot

```python
fig,ax=plt.subplots(figsize=(19,7))
sns.barplot(data=df,x=df.index,y=df.sum(axis=1))
plt.ylabel('Number of Chickenpox cases  per year')
plt.xlabel('YEARS')
plt.title('Barplots showing count of chicken pox for Hungary(2005-2014)')
```

# Visualizing cases for each year using pie chart

```python
def pie_plot(x):
    y=list(df.columns)
    plt.figure(figsize=(15,10))
    vr=0.0
    exp=[vr for x in range(len(y))]
    exp[0]=0.1
    plt.pie(df.loc[x,:],labels=y,autopct='%.1f%%',explode=exp,startangle=90)
    plt.title('Graph showing trend of cases for each county in the year'+str(x))
    plt.show()
```

```python
pie_plot(2005)
```
